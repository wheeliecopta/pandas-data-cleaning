# Let's explore the revenue

## Explore differences in total payment for order and sum price of products in order
Result: Most differences are in the range of shipping costs. In some cases total payment is below sum price, suggesting vouchers or the like.

```python
ol_total_price = orderlines.copy()
```

```python
# Multiply unit price by product quantity in order
ol_total_price["unit_price_total"] = ol_total_price["product_quantity"] * ol_total_price["unit_price"]
```

```python
# Group orderlines by order and get total price of order
ol_grouped = ol_total_price.groupby("id_order", as_index=False)["unit_price_total"].sum()
```

```python
# Create merged dataframe from "orders" and grouped "orderlines"
orders_rev = orders.merge(ol_grouped, how="inner", left_on="order_id", right_on="id_order")
```

```python
# Create column for difference in total price of order and total payment by customer
orders_rev["diff"] = orders_rev["total_paid"] - orders_rev["unit_price_total"]
```

### Let's deal with outliers

```python
# Create first and third quartile for price difference
IQ1 = orders_rev["diff"].quantile(0.25)
IQ3 = orders_rev["diff"].quantile(0.75)
```

```python
# Create list with order_id of outliers according to the 1.5xIQR rule (price difference)
outliers = (orders_rev.loc[
    (orders_rev["diff"] < IQ1 - (IQ3 - IQ1) * 1.5) |
    (orders_rev["diff"] > IQ3 + (IQ3 - IQ1) * 1.5),
    "order_id"]
    .tolist()
)
```

Drop order ids with outliers form all relevant dataframes

```python
# "orders"
orders = orders.loc[~orders["order_id"].isin(outliers)]

# "orderlines"
orderlines = orderlines.loc[~orderlines["id_order"].isin(outliers)]

# "orders_rev"
orders_rev = orders_rev.loc[~orders_rev["order_id"].isin(outliers)]
```

### Explore differences in unit price and base price

Create one dataframe to be able to compare unit and base price

```python
# Create new dataframe, starting with sku and price from "products"
price_diff = products.loc[:, ["sku", "price"]]

# Merge dataframes
price_diff = (orderlines
        .merge(
            price_diff,
            how="left",
            on="sku")
        .merge(
            orders.loc[:, ["created_date", "order_id"]],
            how="left",
            left_on="id_order",
            right_on="order_id")
)

# Only keep relevant columns
price_diff = price_diff.loc[:, "id":"created_date"]
```

Create columns that compare prices

```python
# Create column for price difference between unit and base price
price_diff["price_diff"] = (price_diff["price"] - price_diff["unit_price"]).round(2)

# Create column for the discount in percentage
price_diff["discount"] = (price_diff["price_diff"] * 100 / price_diff["price"]).round(2)
```

Let's look at distribution of price differences

```python
price_diff["price_diff"].describe()
```

```python
price_diff["discount"].describe()
```

### Change unit_price to value of base price if unit_price is higher than price, but not more than 1€

```python
# Create a list with ids of products that have a unit price higher than base price, but difference is not larger than 1€
price_correction = price_diff.loc[(price_diff["price_diff"] < 0) & (price_diff["price_diff"] >= -1), "id"].tolist()


def correct_unit_price(row):
    if row["id"] in price_correction:
        return row["price"]  # Replace "unit_price" with "price" for specific rows
    else:
        return row["unit_price"]

# Apply the function to newly created dataframe
price_diff["unit_price"] = price_diff.apply(correct_unit_price, axis=1)

# But also apply to original "orderlines" dataframe
orderlines["unit_price"] = price_diff["unit_price"].tolist()
```

### Drop all remaining orders that still have a higher unit_price than base price (difference is more than 1€)

```python
# Create list with order ids of corrupted unit prices (meaning higher than base price)
neg_discount = price_diff.loc[price_diff["price_diff"] < 0, "id_order"].unique().tolist()

# Drop all rows with corrupted unit_prices from "orders" (1,616 rows dropped)
orders = orders.loc[~orders["order_id"].isin(neg_discount)]

# Drop all rows with corrupted unit_prices from "orderlines" (2,643 rows dropped)
orderlines = orderlines.loc[~orderlines["id_order"].isin(neg_discount)]

# And from diff_price (also 2,643 rows dropped)
price_diff = price_diff.loc[~price_diff["id_order"].isin(neg_discount)]
```

### Check for outliers in price difference and decide if you want to get rid of orders with these outliers

```python
discountQ1 = price_diff["discount"].quantile(0.25)
discountQ3 = price_diff["discount"].quantile(0.75)

discountIQR = discountQ3 - discountQ1

# Create a list of order ids with outliers in discounts according to the IQRx1.5 rule
discount_outliers = (price_diff.loc[
    (price_diff["discount"] < discountQ1 - discountIQR * 1.5) |
    (price_diff["discount"] > discountQ3 + discountIQR * 1.5),
    "id_order"].tolist()
)

# Drop all rows with discount outliers from "orders" (3,957 rows dropped)
orders = orders.loc[~orders["order_id"].isin(discount_outliers)]

# Drop all rows with discount outliers from "orderlines" (6,464 rows dropped)
orderlines = orderlines.loc[~orderlines["id_order"].isin(discount_outliers)]

# And from diff_price (also 6,464 rows dropped)
price_diff = price_diff.loc[~price_diff["id_order"].isin(discount_outliers)]
```

# Create one merged dataframe with all relevant tables


```python
orders_rev = orders_rev.rename(columns={"diff": "total_price_diff"})
```

```python
# Merge one complete dataframe with all relevant columns
complete_df= (
    orders.merge(
        orderlines,
        how="left", left_on="order_id", right_on="id_order")
    .merge(
        orders_rev.loc[:, ["order_id", "unit_price_total", "total_price_diff"]],
        how="left", on="order_id"
        )
    .merge(
        price_diff.loc[:, ["id_order", "discount"]],
        how="left", on="id_order"
        )
    .merge(
        products.loc[:, ["name", "desc", "type", "sku", "price"]],
        how="left", on="sku"
        )
    )
```

## Make some last adjustments to the complete merged dataframe

```python
# Reduce digit after decimal of "total_price_diff" to 2
complete_df["total_price_diff"] = complete_df["total_price_diff"].round(2)
```

```python
# Drop duplicates
complete_df = complete_df.drop_duplicates()
```

```python
# Change some column names for clarity
complete_df = complete_df.rename(columns={"discount": "discount_%"})
complete_df = complete_df.rename(columns={"id": "product_id"})
```

```python
# Reset index
complete_df = complete_df.reset_index()
```

Add a boolean column: discount yes or no

```python
complete_df["discount_yn"] = True
for index, row in  complete_df.iterrows():
    if row["discount_%"] == 0:
        complete_df.at[index, "discount_yn"] = False
```

```python
# Arrange the columns in a logical order
complete_df = (
    complete_df.loc[:,
                     ["created_date",
                      "product_id",
                      "sku",
                      "name",
                      "desc",
                      "type",
                      "price",
                      "unit_price",
                      "discount_yn",
                      "discount_%",
                      "product_quantity",
                      "unit_price_total",
                      "total_paid",
                      "total_price_diff",
                      "order_id"]
                    ]
    )
```
